#### Exercise 01
Please create a count plot representation of age and charges column values from the provided insurance.csv file.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the insurance dataset
insurance_data = pd.read_csv("insurance.csv")

# Extract the required columns
age_data = insurance_data["age"]
charges_data = insurance_data["charges"]

# Step 1: Count occurrences of each age
age_counts = age_data.value_counts().sort_index()

# Step 2: Aggregate charges per age (optional)
age_charges_mean = insurance_data.groupby("age")["charges"].mean()

# Plotting the count of occurrences for each age
plt.figure(figsize=(12, 6))
plt.bar(age_counts.index, age_counts.values, color='skyblue', alpha=0.8, edgecolor='black')
plt.title("Count Plot: Age Distribution", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Plotting the average charges per age (optional visualization)
plt.figure(figsize=(12, 6))
plt.bar(age_charges_mean.index, age_charges_mean.values, color='orange', alpha=0.8, edgecolor='black')
plt.title("Bar Plot: Average Charges per Age", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Average Charges ($)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### __1.3 adding a thid variable with hue__

#### what is hue?
- a third party variable added to the plots by color
- in this example we will be using "load dataset" function in Seaborn.

```python
# tips is the build-in dataset for learning purpose.
tips = sns.load_dataset("tips")
# show first 5 rows 
tips.head()
# scatter plot the "total bill" with the "tip"
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.show()
```

```python
# hue represents the column name "smoker".

sns.scatterplot(x="total_bill",y="tip",data=tips,hue="smoker",hue_order=["No","Yes"])
plt.show()
```

#### Exercise 02
Please create a scatter plot representation of bmi, charges distribution with age column values from the provided insurance.csv file.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
insurance_data = pd.read_csv("insurance.csv")

# Extract required columns
bmi = insurance_data["bmi"]
charges = insurance_data["charges"]
age = insurance_data["age"]

# Scatter plot
plt.figure(figsize=(12, 8))
scatter = plt.scatter(bmi, charges, c=age, cmap='viridis', alpha=0.7, edgecolor='black')

# Adding color bar to indicate the age
cbar = plt.colorbar(scatter)
cbar.set_label("Age", fontsize=12)

# Plot title and labels
plt.title("Scatter Plot: BMI vs Charges (Colored by Age)", fontsize=16)
plt.xlabel("BMI", fontsize=12)
plt.ylabel("Charges ($)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

#### Exercise 04
Please create a bar plot representation of female and male count in respect to age and charges column values from the provided insurance.csv file.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
insurance_data = pd.read_csv("insurance.csv")

# Group data by age and sex, then calculate the count of rows (observations)
sex_age_counts = insurance_data.groupby(["age", "sex"]).size().unstack()

# Group data by age and sex, then calculate the mean of charges
sex_age_charges = insurance_data.groupby(["age", "sex"])["charges"].mean().unstack()

# Create subplots
fig, axes = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

# Bar plot for counts of males and females
sex_age_counts.plot(kind="bar", stacked=True, ax=axes[0], color=["skyblue", "pink"])
axes[0].set_title("Count of Males and Females by Age", fontsize=16)
axes[0].set_ylabel("Count", fontsize=12)
axes[0].legend(title="Sex", labels=["Male", "Female"])
axes[0].grid(axis="y", linestyle="--", alpha=0.7)

# Bar plot for average charges by age and sex
sex_age_charges.plot(kind="bar", stacked=False, ax=axes[1], color=["skyblue", "pink"])
axes[1].set_title("Average Charges by Age and Sex", fontsize=16)
axes[1].set_ylabel("Average Charges ($)", fontsize=12)
axes[1].set_xlabel("Age", fontsize=12)
axes[1].legend(title="Sex", labels=["Male", "Female"])
axes[1].grid(axis="y", linestyle="--", alpha=0.7)

# Adjust layout
plt.tight_layout()
plt.show()
```

### __3.2 Creating a box plot__
- Shows the distribution of quantitative data 
- See median,spread,skewness,and outliers 
- Facilitates comparisons between groups

```python
# use tips dataset to shwo the "total_bill" box plot accross the "day"
sns.catplot(x="day",y="total_bill",data =tips,kind="box")
```

#### from example above, we need to understand:
- colored box represents 25th to 75 th percentile
- line in the middle is the median
- whiskers give the spread of the distrubution which is 5% and 95% by default.
- the floating points represent the outliers out of 5% and 95%.

#### more over ,there are many ways to customize box plot:
- "sym"= . to remove the outliers.
-  "whis= " can be set to 2.0.instead of 1.5  by default, whiskers extend 1.5 times of interquarile range which is 25% TO 75% range of data.
- "whis" can be set by giving low and higher percentile value like [5,95] or [0,100].


```python
# in this example, the whisker is set to show the lowest and highest value, so no outlier is displayed
import matplotlib.pyplot as pltimport 
import seaborn as sns
sns.catplot(x="time", y="total_bill", data=tips,                
 kind="box",
whis=2.0,
sym= "")
plt.show()
```

### __3.3 point plot__
- show mean of quantitative variable
- vertical line show 95% confidence intervals 

```python
# simple example of ponit plot showing average bill amount between smokers and non-smokers
sns.catplot(x="smoker",y="total_bill",data=tips,kind= "point")
plt.show()
```

#### vertical bar means we are 95% sure that the mean of the total bill amount is within this area if you have another set of data from the same sample space.

![Alt text](images/Screenshot%202022-12-15%20181507.png)

![Alt text](images/Screenshot%202022-12-15%20181613.png)

# by using "join"= "false to disconnects points


```python
# by using "join"= "false to disconnects points
sns.catplot(x="smoker",y="total_bill",
data=tips,
kind= "point",
join = False)
plt.show()
```

```python
# importing numpy Median library to change the mean to median
from numpy import median
sns.catplot(x="smoker",y="total_bill",
data=tips,
kind= "point",
estimator=median)
plt.show()
```

```python
# the confidence interval can be displayed with caps
# the capsize can be  changed by setting numbers.
# or the confidence interval can be turned off by setting 
# errorbar = None

sns.catplot(x="smoker",y="total_bill",
data=tips,
kind= "point",
capsize=0.2)
plt.show()
```

```python
# remove confidence interval
sns.catplot(x="smoker",y="total_bill",
data=tips,
kind= "point",
errorbar=None)
plt.show()
```

### __3.3 Matrix Plots__
-  heatmap is the most common type of matrix plot

#### we used crosstab() in numpy course to align the clusters with the label of incoming data.
#### here the crosstab defines a 2D matrix, which elements are the weekday in a month, and  the means of rentals of that weekday in that month is the value of that element

```python
# crosstab() with "mnth" and "weekday", aggregate function is "mean"
pd.crosstab(bike_data["mnth"],bike_data["weekday"],
values=bike_data["total_rentals"],
aggfunc="mean").round(0)
```

```python
# now we can use heatmap to plot this data.
# the syntax is as below
sns.heatmap(pd.crosstab(bike_data["mnth"],bike_data["weekday"],
values=bike_data["total_rentals"],
aggfunc="mean"))
```

#### the heatmap translates th numerical values in the matrix into a color coded grid,where the color gets lighter as the numbers increase
- so it is easy to see that the busiest days are Saturdays in June and September
- also illustrates that the rental rates are higher in the warmer weather months from May through October

#### next we are going to customize the heatmap to present the most information as clearly as possible.

```python
# assign the crosstab in the previous example to a variable
bike_data_crosstab = pd.crosstab(bike_data["mnth"],bike_data["weekday"],
values=bike_data["total_rentals"],
aggfunc="mean").round(0)

# customize heatmap by passing the crosstab to the function and add other parameters
# "annot = True" turns on the display in the individual cells
# "fmt" set to display one number after decimal point
# cmap changes the color in the cell
# "cbar = False" disable the display of the color bar on the right
# finally we add linewidths to seperate the cells for better view
sns.heatmap(bike_data_crosstab,
annot=True,
fmt=".1f",
cmap="YlGnBu",
cbar = False,
linewidths=0.5)
```

#### further customization is centering a heatmap

```python
# .loc[] specify the indices of the cell you want to center the color for
sns.heatmap(bike_data_crosstab,
annot=True,
fmt=".1f",
cmap="YlGnBu",
cbar = True,
center=bike_data_crosstab.loc[9,6])
```

#### next we will learn how to plotting a __correlation matrix__
- Pandas "corr" function calculates correlations between columns in a dataframe
- the output is a matrix so can be converted to a heatmap

```python
# the color represents the corelation between the columns, ranging from 0 to 1
cols = ['total_rentals', 'temp', 'casual', 'hum', 'windspeed']
sns.heatmap(bike_data[cols].corr(), cmap='YlGnBu')
```

#### from the heatmap, we can find that rentals are highly related to temp and casual values

Make a heatmap/correlation plot for all the values and columns provided in insurance.csv file. Try to find out what is the relation present among the values and features in the insurance.csv file.
